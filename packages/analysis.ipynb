```python
import matplotlib.pyplot as plt
```

```python
company=["apple","oppo","samsung"]
shares=[200,300,400]
```

```python
plt.bar(company,shares)
plt.xlabel("company name")
plt.ylabel("shares")
plt.title("Distribution of shares of different companies")
```

```python
plt.pie(shares,labels=["apple","oppo","samsung"],colors=["red","blue","green"])
plt.axis("off")
plt.title("Distribution of shares")
```

```python
plt.line(shares,company)
```
