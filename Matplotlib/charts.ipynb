```python
import matplotlib.pyplot as plt
import numpy as np
 
company=["GOOGLE","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
profit=[40,2,36,12]

plt.bar(company,revenue , label='revenue')
plt.bar(company,profit , label='profit')
plt.legend()
```

```python
plt.barh(company,revenue , label='revenue') # barh comamnd plot the horizontal bar chart 
plt.barh(company,profit , label='profit')
plt.legend()
```
