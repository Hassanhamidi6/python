```python
import matplotlib.pyplot as plt 
```

```python
x=["hassan","arham","ali","rahdain","taimoor"]
y=[80,75,45,70,90]
plt.xlabel("Student names")
plt.ylabel("Student marks")
plt.title("Students result")
plt.plot(x,y ,  color="black" , linewidth=6 , linestyle="solid", markersize=10)
```

```python
x=["hassan","arham","ali","rahdain","taimoor"]       # alpha is use for the transparencyof the line 
y=[80,75,45,70,90]
plt.plot(x,y ,  color="black", marker='D', markersize=10, alpha=0.5) 
```
