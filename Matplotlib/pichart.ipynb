```python
import matplotlib.pyplot as plt
```

```python
exp_values=[1400, 600, 300, 410, 250]
exp_labels=["Home-rent", "Food", "Internet bill", "Car", "Other utilities"]
```

```python
plt.axis("equal")   # it is use to make a oerfect circle of pie chart
plt.pie(exp_values,labels=exp_labels  , radius= 1, autopct='%0.0f%%' , 
        explode=[0,0.2,0,0,0], startangle=360)      
# radius is use to change the size of the pie   
#autopct is use to write percentages on the pie chart
#explode is use to come out the pie slice
# we can also set the angle by using startangle command

plt.show()    # it will ignore the extra text which will come after running it
```
