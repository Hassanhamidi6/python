```python
import pandas as pd 
import numpy as np
df=pd.read_csv("weather.csv")
df
```

```python
new_df=df.replace({
    "temperature":-99999,
    "windspeed":-99999,
    "event": "no event"
},np.nan)
new_df
```

```python
new_df=df.replace({
    -99999:np.nan,
    "no event": 'sunny'
})
new_df
```

```python
new_df=df.replace({
    "temperature": '[A-Za-z]',
    "windspeed":'[A-Za-z]'
},'',regex=True)        #any charcter between a to z will be replace by the blank value 
new_df                  #here regex is use to change the origenal dataframe
```
